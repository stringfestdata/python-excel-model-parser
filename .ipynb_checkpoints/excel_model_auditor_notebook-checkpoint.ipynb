# Building a Simple Excel Model Auditor with Python

In this notebook we build a lightweight Excel auditing tool using Python.

The goal is to scan an Excel workbook and identify common spreadsheet risks such as:

- Hardcoded numbers inside formulas
- Constants appearing in calculation sheets
- Links to external workbooks
- Inconsistent formulas across a column

This is not a full spreadsheet auditing system. It is a **minimum viable example** showing how Python can inspect Excel models programmatically.

**Download the exercise files and place the workbook in the same folder as this notebook before running the code.**

## Step 1 — Import Libraries

We use:

- `openpyxl` to read Excel workbooks
- `re` (regular expressions) to inspect formulas

Think of `openpyxl` as a way to let Python *look inside* an Excel file.

````python
import openpyxl
import re
````

## Step 2 — Load the Workbook

`data_only=False` ensures that we read the **formulas themselves**, not just the calculated values.

This is important because we want to inspect the structure of the model.

````python
file = "demo_financial_model.xlsx"
wb = openpyxl.load_workbook(file, data_only=False)
````

## Step 3 — Define Audit Rules

Here we define some simple modeling rules.

For example, many financial models follow a pattern where:

- Input sheets contain constants
- Calculation sheets contain formulas

We'll flag violations of this structure.

````python
findings = []

# Pattern for Excel cell references
cell_ref_pattern = r"[A-Za-z_]+!\$?[A-Za-z]+\$?\d+|\$?[A-Za-z]+\$?\d+"

# Sheets that are allowed to contain constants
input_sheets = ["Inputs"]
````

## Step 4 — Scan the Workbook

Excel files have a simple hierarchy:

**Workbook → Sheets → Rows → Cells**

We loop through every cell and analyze its contents.

````python
for sheet in wb.sheetnames:

    ws = wb[sheet]

    for row in ws.iter_rows():
        for cell in row:

            location = f"{sheet}!{cell.coordinate}"
````

## Step 5 — Inspect Formulas

We look for two risks:

1. Hardcoded numbers inside formulas
2. External workbook references

Hardcoded numbers can make models fragile because the logic is hidden inside formulas.

````python
            if cell.data_type == "f":

                formula = cell.value

                cleaned_formula = re.sub(cell_ref_pattern, "", formula)
                numbers = re.findall(r"\b\d+(?:\.\d+)?\b", cleaned_formula)

                for n in numbers:

                    value = float(n)

                    if value in [0, 1]:
                        continue

                    findings.append({
                        "issue": "Hardcoded number inside formula",
                        "cell": location,
                        "formula": formula
                    })

                    break

                if "[" in formula and "]" in formula:

                    findings.append({
                        "issue": "External workbook link",
                        "cell": location,
                        "formula": formula
                    })
````

## Step 6 — Detect Constants in Calculation Sheets

If a cell contains a number but appears outside the Inputs sheet, we flag it.

````python
            elif cell.data_type == "n":

                if sheet not in input_sheets:

                    if cell.value is not None:

                        findings.append({
                            "issue": "Hardcoded constant cell",
                            "cell": location,
                            "value": cell.value
                        })
````

## Step 7 — Detect Inconsistent Formulas

One of the most common spreadsheet errors occurs when formulas are inconsistent across a column.

Example:

```
=B2*C2
=B3*C3
=B4*C4
=B5*D5   ← mistake
```

We compare formulas down a column to see if they follow the same pattern.

````python
for sheet in wb.sheetnames:

    ws = wb[sheet]

    for col in ws.iter_cols():

        formulas = []

        for cell in col:
            if cell.data_type == "f":
                formulas.append((cell.coordinate, cell.value))

        if len(formulas) > 1:

            base_formula = formulas[0][1]

            for coord, f in formulas[1:]:

                if f != base_formula:

                    findings.append({
                        "issue": "Inconsistent formula pattern",
                        "cell": f"{sheet}!{coord}",
                        "formula": f
                    })
````

## Step 8 — Display Results

Finally we print the audit findings.

````python
print("\nAUDIT FINDINGS\n")

for f in findings:

    print(f["issue"])

    if "formula" in f:
        print("Formula:", f["formula"])

    if "value" in f:
        print("Value:", f["value"])

    print("Cell:", f["cell"])
    print()
````
