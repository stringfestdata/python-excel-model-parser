```python
import openpyxl
import re

file = "demo_financial_model.xlsx"

wb = openpyxl.load_workbook(file, data_only=False)

findings = []

# Pattern for Excel cell references
cell_ref_pattern = r"[A-Za-z_]+!\$?[A-Za-z]+\$?\d+|\$?[A-Za-z]+\$?\d+"

# Sheets that are allowed to contain constants
input_sheets = ["Inputs"]

for sheet in wb.sheetnames:

    ws = wb[sheet]

    for row in ws.iter_rows():
        for cell in row:

            location = f"{sheet}!{cell.coordinate}"

            # -------------------------
            # 1. Check formulas
            # -------------------------
            if cell.data_type == "f":

                formula = cell.value

                # Remove cell references
                cleaned_formula = re.sub(cell_ref_pattern, "", formula)

                # Find numeric constants
                numbers = re.findall(r"\b\d+(?:\.\d+)?\b", cleaned_formula)

                for n in numbers:

                    value = float(n)

                    # Ignore structural constants
                    if value in [0, 1]:
                        continue

                    findings.append({
                        "issue": "Hardcoded number inside formula",
                        "cell": location,
                        "formula": formula
                    })

                    break

                # Detect external links
                if "[" in formula and "]" in formula:

                    findings.append({
                        "issue": "External workbook link",
                        "cell": location,
                        "formula": formula
                    })

            # -------------------------
            # 2. Detect constants in calculation sheets
            # -------------------------
            elif cell.data_type == "n":

                if sheet not in input_sheets:

                    if cell.value is not None:

                        findings.append({
                            "issue": "Hardcoded constant cell",
                            "cell": location,
                            "value": cell.value
                        })


# -------------------------
# Print Results
# -------------------------

print("\nAUDIT FINDINGS\n")

for f in findings:

    print(f["issue"])

    if "formula" in f:
        print("Formula:", f["formula"])

    if "value" in f:
        print("Value:", f["value"])

    print("Cell:", f["cell"])
    print()
```

```python
import openpyxl
import networkx as nx
import re
import matplotlib.pyplot as plt

file = "demo_financial_model.xlsx"

wb = openpyxl.load_workbook(file, data_only=False)

G = nx.DiGraph()

# regex to capture cell references
cell_pattern = r"[A-Za-z_]+!\$?[A-Za-z]+\$?\d+|\$?[A-Za-z]+\$?\d+"

for sheet in wb.sheetnames:

    ws = wb[sheet]

    for row in ws.iter_rows():

        for cell in row:

            if cell.data_type == "f":

                target = f"{sheet}!{cell.coordinate}"

                formula = cell.value

                refs = re.findall(cell_pattern, formula)

                for ref in refs:

                    # if reference doesn't include sheet, assume same sheet
                    if "!" not in ref:
                        ref = f"{sheet}!{ref}"

                    ref = ref.replace("$","")

                    G.add_edge(ref, target)

print("Nodes:", len(G.nodes))
print("Edges:", len(G.edges))
```

```python
plt.figure(figsize=(12,8))

pos = nx.spring_layout(G)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=2000,
    node_color="lightblue",
    font_size=8,
    arrows=True
)

plt.show()
```
